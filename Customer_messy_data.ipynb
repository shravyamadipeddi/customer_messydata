```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
```

# Task 1 — Clean the Data

```python
df.describe()
```

## Impute missing values in age with the median and in city with the mode

```python
df['age'] = df['age'].fillna(df['age'].median())
df['city'] = df['city'].fillna(df['city'].mode()[0])

df
```

# Task 2 — Encode Categorical Columns

```python
from sklearn.preprocessing import LabelEncoder
```

```python
print(df.columns)
```

## Apply One-Hot Encoding to city

```python
if 'city' in df.columns:
    df = pd.get_dummies(df, columns=['city'], prefix='city')
df
```

## Apply Label Encoding to gender

```python
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

print("Encoded DataFrame:\n", df)
```

# Task 3 — Scale Numerical Features

### Min-Max Scaling (using MinMaxScaler)
### Standardisation (using StandardScaler)

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler_minmax = MinMaxScaler()
scaler_standard = StandardScaler()

df_minmax = df.copy()
df_minmax[['age', 'annual_income']] = scaler_minmax.fit_transform(df_minmax[['age', 'annual_income']])

df_standard = df.copy()
df_standard[['age', 'annual_income']] = scaler_standard.fit_transform(df_standard[['age', 'annual_income']])

print("\nMin-Max Scaled DataFrame:\n", df_minmax)
print("\nStandardised DataFrame:\n", df_standard)
```

### **Min‑Max Scaling:** rescales values into a fixed range [0,1]. Useful when you want bounded features (e.g., neural networks).

### **Standardisation:** centers data to mean 0, variance 1. Better when features follow a Gaussian distribution or when outliers exist.
